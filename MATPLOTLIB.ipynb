```python
var=pd.DataFrame({'normal':np.random.normal(size=10),'gamma':np.random.normal(1,size=10),'poisson':np.random.poisson(size=10)})
print(var)
var.cumsum(0).plot()
```

```python
var.cumsum(0).plot(subplots=True)
```

```python
var.cumsum(0).plot(secondary_y='normal')
```

```python
var.cumsum(0).plot(secondary_y='normal')
```

```python
fig,axes = plt.subplots(nrows=3, ncols=1, figsize=(3, 10))
for i,j in enumerate(['normal','gamma','poisson']):
    var[j].cumsum(0).plot(ax=axes[i],title=j)
axes[0].set_ylabel('cumulative sum')
```

```python
mean = [5, 2]
cov = [[3, 1], [2, 0.5]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x, y, bins=20, cmap='Reds')
cb = plt.colorbar()
```

```python
plt.hexbin(x, y, gridsize=20, cmap='Reds')
cb = plt.colorbar()
```
